# Popularity based Recommendation System


```python
ratings_with_name = df_reviews.merge(df_books, on = 'isbn')
# ratings_with_name
```

```python

num_rating_df = ratings_with_name.groupby('title').count()['rating'].reset_index()
num_rating_df.rename(columns={'rating':'num_ratings'}, inplace = True)
num_rating_df.head()
```

```python
#Average Rating
# avg_rating_df = ratings_with_name.groupby('title').mean()['rating'].reset_index()
# avg_rating_df.rename(columns={'rating':'avg_rating'}, inplace = True)
# avg_rating_df 
# Convert 'rating' column to numeric
ratings_with_name['rating'] = pd.to_numeric(ratings_with_name['rating'], errors='coerce')

# Average Rating
avg_rating_df = ratings_with_name.groupby('title')['rating'].mean().reset_index()
avg_rating_df.rename(columns={'rating':'avg_rating'}, inplace=True)
```

```python
#merge average rating and number rating
popular_df = num_rating_df.merge(avg_rating_df, on = 'title')
popular_df.head()
```

```python
#shows popular book whose num rating>3
popular_df = popular_df[popular_df['num_ratings'] >=2].sort_values('avg_rating',ascending =False)
# popular_df
```

```python
# popular_df.merge(df_books, on='title')
```

```python
popular_df = popular_df.merge(df_books, on='title').drop_duplicates('title')[['title', 'num_ratings', 'avg_rating']]
```

## Collaborative Filtering

```python
# ratings_with_name.groupby('created_by_id').count()['rating'] > 10
```

```python
# ratings_with_name.head()
# coll_fil = pd.DataFrame(ratings_with_name)
# coll_fil
```

```python
#those users who have rated more than 10 books
x = ratings_with_name.groupby('created_by_id').count()['rating'] > 10
read_users = x[x].index
# read_users
```

```python
#filtered on the basis of rating
filtered_rating = ratings_with_name[ratings_with_name['created_by_id'].isin(read_users)]
# filtered_rating
```

```python
#those books that have at least 5 number of rating
y = filtered_rating.groupby('title').count()['rating'] >= 5
famous_books = y[y].index
# famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_books)]
# final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'title', columns = 'created_by_id', values = 'rating')
# final_ratings.to_frame().reset_index().pivot_table(index='title')
```

```python
pt.fillna(0, inplace = True)
```

```python
# pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
#suggest five books
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = df_books[df_books['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['author'].values))
        item.extend(list(temp_df.drop_duplicates('title')['image'].values))
        
        data.append(item)
    
    return data
```

```python
# sorted(list(enumerate(similarity_scores[0])),key=lambda x:x[1],reverse=True)
```

```python
recommend('C Programming An Intorduction')
```
