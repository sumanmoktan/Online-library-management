```python
import os
import django
import pandas as pd
import numpy as np

```

```python
from books.models import Pdf_Info, Review, Genre
from auth_users.models import Profile
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from asgiref.sync import sync_to_async

@sync_to_async
def get_books():
    return list(Pdf_Info.objects.all().values())

books = await get_books()
df_books = pd.DataFrame(books)
df_books.head()
```

```python
from asgiref.sync import sync_to_async

@sync_to_async
def get_users():
    return list(Profile.objects.all().values())

users = await get_users()
df_users = pd.DataFrame(users)
df_users.head()
```

```python
@sync_to_async
def get_reviews():
    return list(Review.objects.all().values())

reviews = await get_reviews()
df_reviews = pd.DataFrame(reviews)
df_reviews.head()
```

```python
@sync_to_async
def get_reviews():
    return list(Genre.objects.all().values())

genre = await get_reviews()
df_genre = pd.DataFrame(genre)
df_genre.head()
```

```python
ratings_with_name = df_reviews.merge(df_books, on = 'isbn')
```

```python
# merge books dataframe with genre dataframe
# books_with_genre = df_genre.merge(df_books)
# books_with_genre.head()
```

```python
#those users who have rated more than 10 books
x = ratings_with_name.groupby('created_by_id').count()['rating'] > 10
read_users = x[x].index
read_users
```

```python
#filtered on the basis of rating
filtered_rating = ratings_with_name[ratings_with_name['created_by_id'].isin(read_users)]
```

```python
#those books that have at least 5 number of rating
y = filtered_rating.groupby('title').count()['rating'] >= 5
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index = 'title', columns = 'created_by_id', values = 'rating')
pt
```

```python
pt.fillna(0, inplace = True)
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # Fetch index of searched book
    index = np.where(pt.index.str.lower().str.contains(book_name.lower()))[0][0]

    # Get similarity scores for all books
    similarity_scores = cosine_similarity(pt)

    # Get indices of similar books with same genre
    similar_items = []
    for i in range(similarity_scores.shape[0]):
        # Check if book has same genre as searched book
        if df_books.loc[df_books['title'] == pt.index[i], 'genre_id'].iloc[0] == df_books.loc[df_books['title'] == pt.index[index], 'genre_id'].iloc[0]:
            # Append index and similarity score to list
            if i != index:
                similar_items.append((i, similarity_scores[index][i]))

    # Sort by similarity score
    similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)[:5]

    # Get book information and append to list
    data = []
    for i in similar_items:
        item = []
        temp_df = df_books[df_books['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['author'].values))
        item.extend(list(temp_df.drop_duplicates('title')['image'].values))
        data.append(item)

    return data
```
