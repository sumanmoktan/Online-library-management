```python
ratings_with_name = df_reviews.merge(df_books, on = 'isbn')
ratings_with_name
```

```python
# Aggregrate  By book
agg_ratings = ratings_with_name.groupby('title').agg(mean_rating =('rating', 'mean'), number_of_rating = ('rating', 'count')).reset_index()

# Keep the books with over 3 ratings
agg_ratings_GT = agg_ratings[agg_ratings['number_of_rating']>3].sort_values(by = 'number_of_rating')
agg_ratings_GT.head(10)
# avg_rating_df.head()
```

```python
df_gt100 = pd.merge(ratings_with_name, agg_ratings_GT[['title']], on = 'title', how = 'inner')
```

```python
df_gt100.info()
```

```python
matrix = df_gt100.pivot_table(index = 'created_by_id', columns = ['title','description','isbn','author', 'image','book_pdf'], values = 'rating')
matrix
import pickle
pickle.dump(matrix,open('matrix.pkl','wb'))
```

```python
 #Normalization User_Item Matrix
matrix_norm = matrix.subtract(matrix.mean(axis = 1), axis = 'rows')
matrix_norm.head()
```

```python
#Use Similarity Matrix using Pearson Correlation
user_similarity = matrix_norm.T.corr()
user_similarity.head()
```

```python
#User Similarity matrix using cosine similarity
user_similarity_cosine = cosine_similarity(matrix_norm.fillna(0))
user_similarity_cosine
```

```python
#Picke a User ID
picked_userid =2

#remove picked user id from the condidate list
user_similarity.drop(index = picked_userid, inplace = True)

#Take a look at the data
user_similarity.head()
```

```python
 #number of similar users
n = 10

#User similarity threashold
user_similarity_threshold = 0.1

#Get top n similar users
similar_users = user_similarity[user_similarity[picked_userid] > user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]
print(f'The similar users for user {picked_userid} are', similar_users)
```

```python
# A dictionary to store item scores
item_score = {}

# Loop through items
for i in matrix.columns:
    # Get the rating for book i
    book_rating = matrix[i]
    # Create variables to store the score and count
    total = 0
    count = 0
    # Loop through similar users
    for u in similar_users.index:
        # If the user has rated the book
        if pd.isna(book_rating[u]) == False:
            # Calculate the score
            score = similar_users[u] * book_rating[u]
            # Add the score to the total
            total += score
            # Increment the count
            count += 1
    # Calculate the average score for the book
    item_score[i] = {
        'title': i[0],
        'description': i[1],
        'isbn': i[2],
        'author': i[3],
        'image': i[4],
        'book_pdf': i[5],
        'book_score': total / count if count > 0 else 0,
    }

# Convert the dictionary to a dataframe
item_score_df = pd.DataFrame.from_dict(item_score, orient='index')

# Sort the books by score
ranked_item_score = item_score_df.sort_values(by='book_score', ascending=False)

# Select the top m books
m = 10
top_books = ranked_item_score.head(m)
ranked_item_score.head(m)
```
