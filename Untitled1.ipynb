```python
from books.models import Review, Pdf_Info
from auth_users.models import Profile
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```python
from asgiref.sync import sync_to_async

@sync_to_async
def get_books():
    return list(Pdf_Info.objects.all().values())

books = await get_books()
df_books = pd.DataFrame(books)
df_books.head()
```

```python
@sync_to_async
def get_reviews():
    return list(Review.objects.all().values())

reviews = await get_reviews()
df_reviews = pd.DataFrame(reviews)
df_reviews.head()
```

```python
@sync_to_async
def get_users():
    return list(Profile.objects.all().values())

users = await get_users()
df_users = pd.DataFrame(users)
df_users.head()
```

```python
ratings_with_name = df_reviews.merge(df_books, on = 'isbn')
ratings_with_name
```

```python
# Aggregrate  By book
agg_ratings = ratings_with_name.groupby('title').agg(mean_rating =('rating', 'mean'), number_of_rating = ('rating', 'count')).reset_index()

# Keep the movies with over 100 ratings
agg_ratings_GT = agg_ratings[agg_ratings['number_of_rating']>2].sort_values(by = 'number_of_rating')
agg_ratings_GT.head(10)
# avg_rating_df.head()
```

```python
df_gt100 = pd.merge(ratings_with_name, agg_ratings_GT[['title']], on = 'title', how = 'inner')
```

```python
df_gt100.info()
```

```python
matrix = df_gt100.pivot_table(index = 'created_by_id', columns = ['title','description','isbn','author'], values = 'rating')
matrix
```

```python
 #Normalization User_Item Matrix
matrix_norm = matrix.subtract(matrix.mean(axis = 1), axis = 'rows')
matrix_norm.head()
```

```python
#Use Similarity Matrix using Pearson Correlation
user_similarity = matrix_norm.T.corr()
user_similarity.head()
```

```python
#User Similarity matrix using cosine similarity
user_similarity_cosine = cosine_similarity(matrix_norm.fillna(0))
user_similarity_cosine
```

```python
# user_similarity =user_similarity_cosine.T.corr()
# user_similarity.head()
```

```python
#Picke a User ID
picked_userid =1

#remove picked user id from the condidate list
user_similarity_cosine.drop(index = picked_userid, inplace = True)

#Take a look at the data
user_similarity_cosine.head()
```

```python
 #number of similar users
n = 10

#User similarity threashold
user_similarity_threshold = 0.1

#Get top n similar users
similar_users = user_similarity[user_similarity[picked_userid] > user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]
print(f'The similar users for user {picked_userid} are', similar_users)
```

```python
# Books that Targeted User has Read
picked_userid_watched = matrix_norm[matrix_norm.index == picked_userid].dropna(axis = 1, how = 'all')
picked_userid_watched 
```

```python
# Books that similar Users Watched.remove movies that none of the similar users have watched
# similar_users_books = matrix_norm[matrix_norm.index.isin(similar_users.index)].dropna(axis = 1, how = 'all')
similar_users_books = matrix_norm[matrix_norm.index.isin(similar_users.index)].loc[:, :]
similar_users_books
```

```python
#Remove the watched book from the movie list
similar_users_books.drop(picked_userid_watched.columns, axis = 1, inplace = True, errors = 'ignore')


#take a look at the data
similar_users_books
```

```python
#A Dictionary to library item scores
item_score = {}

#loop through items
for i in similar_users_books.columns:
    print(i ,end='\n')
    #get the rating for book i
    book_rating = similar_users_books[i]
    #create a variable to store the score
    total = 0
    #create a variable to store the score
    count = 0
    #loop through similar users
    for u in similar_users.index:
        #if the movie has rating
        if pd.isna(book_rating[u]) == False:
            #score is the sum of user similarity score multiply the book rating
            score = similar_users[u] * book_rating[u]
            #Add the score to the total score for the movie so far
            total += score
            #add 1 to the count
            count += 1
    #get the average score for the item
    item_score[i] = total / count
        
#convert dictionary pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns = ['title', 'book_score',])

#sort the book by score
ranked_item_score = item_score.sort_values(by = 'book_score', ascending = False)


#select top m books
m = 10
ranked_item_score.head(m)
```

```python
# A dictionary to store item scores
item_score = {}

# Loop through items
for i in matrix.columns:
    # Get the rating for book i
    book_rating = matrix[i]
    # Create variables to store the score and count
    total = 0
    count = 0
    # Loop through similar users
    for u in similar_users.index:
        # If the user has rated the book
        if pd.isna(book_rating[u]) == False:
            # Calculate the score
            score = similar_users[u] * book_rating[u]
            # Add the score to the total
            total += score
            # Increment the count
            count += 1
    # Calculate the average score for the book
    item_score[i] = {
        'title': i[0],
        'description': i[1],
        'isbn': i[2],
        'author': i[3],
        'book_score': total / count if count > 0 else 0,
    }

# Convert the dictionary to a dataframe
item_score_df = pd.DataFrame.from_dict(item_score, orient='index')

# Sort the books by score
ranked_item_score = item_score_df.sort_values(by='book_score', ascending=False)

# Select the top m books
m = 10
top_books = ranked_item_score.head(m)
ranked_item_score.head(m)
```
