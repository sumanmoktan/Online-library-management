```python
def recommend(book_name):
    # Fetch index of searched book
    index = np.where(pt.index.str.lower().str.contains(book_name.lower()))[0][0]

    # Get similarity scores for all books
    similarity_scores = cosine_similarity(pt)

    # Get indices of similar books with same genre
    similar_items = []
    for i in range(similarity_scores.shape[0]):
        # Check if book has same genre as searched book
        if df_books.loc[df_books['title'] == pt.index[i], 'genre_id'].iloc[0] == df_books.loc[df_books['title'] == pt.index[index], 'genre_id'].iloc[0]:
            # Append index and similarity score to list
            if i != index:
                similar_items.append((i, similarity_scores[index][i]))

    # Sort by similarity score
    similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)[:9]

    # Get book information and append to list
    data = []
    for i in similar_items:
        item = []
        temp_df = df_books[df_books['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['author'].values))
        item.extend(list(temp_df.drop_duplicates('title')['image'].values))
        data.append(item)

    return data
```

```python
def recommend(book_name):
    # Fetch index of searched book
    match_index = np.where(pt.index.str.lower().str.contains(book_name.lower(), case=False))[0]
    if len(match_index) == 0:
        return []

    # Get similarity scores for all books
    similarity_scores = cosine_similarity(pt)

    # Get indices of similar books with same genre
    similar_books = []
    for i in range(similarity_scores.shape[0]):
        if i in match_index:
            # Check if book has same genre as searched book
            if df_books.loc[df_books['title'] == pt.index[i], 'genre_id'].iloc[0] == df_books.loc[df_books['title'] == pt.index[match_index[-1]], 'genre_id'].iloc[0]:
                # Append title to list
                if i != match_index[-1]:
                    similar_books.append(pt.index[i])

    # Get all reviews for similar books
    similar_reviews = df_reviews[df_reviews['book_id'].isin(similar_books)]

    # Group reviews by book title
    grouped_reviews = similar_reviews.groupby('book_id')

    # Create dictionary to store results
    result = {}

    # Iterate through groups and retrieve reviews
    for title, group in grouped_reviews:
        temp_pdf_info = df_books[df_books['title'] == title]
        if len(temp_pdf_info) == 0:
            continue
        item = {}
        item['title'] = temp_pdf_info.iloc[0]['title']
        item['author'] = temp_pdf_info.iloc[0]['author']
        item['image'] = temp_pdf_info.iloc[0]['image']
        item['description'] = temp_pdf_info.iloc[0]['description']
        item['isbn'] = temp_pdf_info.iloc[0]['isbn']
        item['published_year'] = temp_pdf_info.iloc[0]['published_year']
        item['book_pdf'] = temp_pdf_info.iloc[0]['book_pdf']
        item['is_premium'] = temp_pdf_info.iloc[0]['is_premium']
        item['reviews'] = []
        for review in group.itertuples():
            temp_review = {}
            temp_review['rating'] = review.rating
            temp_review['content'] = review.content
            temp_review['created_at'] = review.created_at
            temp_review['created_by'] = review.created_by.username
            item['reviews'].append(temp_review)
        result[title] = item

    return result.values()
```
